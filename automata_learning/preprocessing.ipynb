```python
import numpy as np
from sklearn.preprocessing import KBinsDiscretizer
total_features=[]
for i in range(1,76):
    features=np.load('traces/trace%d_features.npy' %(i))
    features=np.reshape(features, (features.shape[0],features.shape[-1]))
    total_features.append(features)
total_features=np.concatenate(total_features,axis=0)
total_features.shape
```

```python
from sklearn.preprocessing import KBinsDiscretizer
est = KBinsDiscretizer(n_bins=5, encode='ordinal', strategy='uniform')
est.fit(total_features)
```

```python
!rm traces.psm
for i in range(1,76):
    features=np.load('traces/trace%d_features.npy' %(i))
    features=np.reshape(features, (features.shape[0],features.shape[-1]))
    Xt=est.transform(features).astype('int32')
    rewards=np.load('traces/trace%d_rewards.npy' %(i))
    rewards=rewards.reshape(-1).astype('int32')
    with open("traces.psm", "a") as outfile:
        outfile.write("hmm1(%s,%s).\n" %(str(Xt.tolist()),str(rewards.tolist())))
```
